# A Simple Trading Algorithim

---

### Import Libraries and Dependencies

```python
# Imports
import pandas as pd
import numpy as np
%matplotlib inline
```

### Read in TSLA Dates and Closing Prices as Pandas DataFrame

```python
# Hard-coded dates and closing prices for TSLA
tsla_df = pd.DataFrame(
    {"close": [839.81, 872.79, 854.69, 849.99, 852.23, 863.42, 849.46, 804.82, 811.66, 816.12, 796.22, 798.15, 787.38, 781.30, 714.14, 725.13]}
)

# Set the index as datetime objects starting from 2021-02-01 and onwards for 16 periods (but only business days)
tsla_df.index = pd.bdate_range(start='2021-02-01', periods=16)

# Review the DataFrame
tsla_df
```

### Visualize TSLA Dates amd Closing Prices

```python
# Use the `plot` function to plot the closing prices
tsla_df.plot()
```

### Create a Simple Trading Algorithm

```python
# Initialize `side` column for buys and sells
tsla_df['side'] = np.nan

# Initialize variable to hold previous price
previous_price = 0

# Loop through the Pandas DataFrame and initiate a... 
# buy if the previous price is 0, in other words, buy on the first day
# buy if the current day price is less than the previous day price, and
# sell if the current day price is greater than the previous day price, and
# hold if the current day price is equal to the previous day price
for index, row in tsla_df.iterrows():
    if previous_price == 0:
        tsla_df.loc[index, "side"] = "buy"
    elif row["close"] < previous_price:
        tsla_df.loc[index, "side"] = "buy"
    elif row["close"] > previous_price:
        tsla_df.loc[index, "side"] = "sell"
    else:
        tsla_df.loc[index, "side"] = "hold"

    previous_price = row["close"]

    if index == tsla_df.index[-1]:
        tsla_df.loc[index, "side"] = "sell"

# Print the DataFrame
tsla_df
```

### Modify Trading Algorithm to Include Profit/Loss Metrics for a Long-position Strategy

```python
# Initialize trade_type column to track buys and sells
tsla_df["side"] = np.nan

# Initialize a cost/proceeds column for recording trade metrics
tsla_df["cost/proceeds"] = np.nan

# Initialize share size and accumulated shares
share_size = 100
accumulated_shares = 0

# Initialize variable to hold previous price
previous_price = 0

# Loop through the Pandas DataFrame and initiate a trade at each iteration 
for index, row in tsla_df.iterrows():
    
    # initiate a buy on the first trading day
    if previous_price == 0:
        tsla_df.loc[index, "side"] = "buy"
        tsla_df.loc[index, "cost/proceeds"] = -(row["close"] * share_size)
        accumulated_shares += share_size
        
    # buy if the current day's price is less than the previous day's price
    elif row["close"] < previous_price:
        tsla_df.loc[index, "side"] = "buy"
        tsla_df.loc[index, "cost/proceeds"] = -(row["close"] * share_size)
        accumulated_shares += share_size
        
    # sell if the current day's price is greater than the previous day's price
    elif row["close"] > previous_price:
        tsla_df.loc[index, "side"] = "sell" 
        tsla_df.loc[index, "cost/proceeds"] = (row["close"] * accumulated_shares)
        accumulated_shares = 0
        
    # hold if the current day's price is equal to the previous day's price
    else:
        tsla_df.loc[index, "side"] = "hold"

    # update the previous_price to the current row's price
    previous_price = row["close"]
    

# Review the DataFrame
tsla_df
```

```python
# Determine the net cost/proceeds of the algorithm 
cp = tsla_df["cost/proceeds"].sum()
cp
```

## Modify the Algorithm for a Short-position trading strategy

```python
# Re-create the original DataFrame

# Hard-coded dates and closing prices for TSLA
tsla_df = pd.DataFrame(
    {"close": [839.81, 872.79, 854.69, 849.99, 852.23, 863.42, 849.46, 804.82, 811.66, 816.12, 796.22, 798.15, 787.38, 781.30, 714.14, 725.13]}
)

# Set the index as datetime objects starting from 2021-02-01 and onwards for 16 periods (but only business days)
tsla_df.index = pd.bdate_range(start='2021-02-01', periods=16)

# Review the DataFrame
tsla_df
```

```python
# Initialize trade_type column to track buys and sells
tsla_df["side"] = np.nan

# Initialize a cost/proceeds column for recording trade metrics
tsla_df["cost/proceeds"] = np.nan

# Initialize share size and accumulated shares
share_size = -100
accumulated_shares = 0

# Initialize variable to hold previous price
previous_price = 0

# Loop through the Pandas DataFrame and initiate a trade at each iteration 
for index, row in tsla_df.iterrows():
    
    # initiate the trade with a short sale on the first day
    if previous_price == 0:
        tsla_df.loc[index, "side"] = "sell_short"
        tsla_df.loc[index, "cost/proceeds"] = -(row["close"] * share_size)
        accumulated_shares += share_size
        tsla_df.loc[index, "shares"] = accumulated_shares
        
    # buy to cover if the current day's price is less than the previous day's price
    elif row["close"] < previous_price:
        tsla_df.loc[index, "side"] = "buy to cover"
        tsla_df.loc[index, "cost/proceeds"] = (row["close"] * accumulated_shares)
        accumulated_shares = 0
        tsla_df.loc[index, "shares"] = accumulated_shares
        
    # sell short if the current day's price is greater than the previous day's price
    elif row["close"] > previous_price:
        tsla_df.loc[index, "side"] = "sell_short"
        tsla_df.loc[index, "cost/proceeds"] = -(row["close"] * share_size)
        accumulated_shares += share_size
        tsla_df.loc[index, "shares"] = accumulated_shares
    
    # hold if the current day's price is equal to the previous day's price
    else:
        tsla_df.loc[index, "side"] = "hold"

    # update the previous_price to the current row's price
    previous_price = row["close"]
    
    # if the index is the last index of the DataFrame, close the position
    if index == tsla_df.index[-1]:
        tsla_df.loc[index, "side"] = "close_position"
        if accumulated_shares < 0:
            accumulated_shares = 0; 
            tsla_df.loc[index, "cost/proceeds"] = row["close"] * accumulated_shares
            tsla_df.loc[index, "shares"] = accumulated_shares

# Review the DataFrame
tsla_df
```

```python
# Determine the net cost/proceeds of the algorithm 
cp = tsla_df["cost/proceeds"].sum()
cp
```
